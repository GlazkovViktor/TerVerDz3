## Задание 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('kc_house_data.csv')
df
```

### Задача 1.2

```python
plt.figure(figsize=(8, 8))
plt.hist(df['price'], bins = 100)
plt.title('Распределение цен')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

Вывод:
1. Максимальное количество недвижимости продается по цене до ~900 000, с увеличением цены, уменьшается количество домов 

### Задание 1.3

```python
plt.figure(figsize=(6, 6))
plt.hist(df['sqft_living'], bins = 1000)
plt.title('Распределение квадратуры жилой')
plt.xlabel('Площадь')
plt.ylabel('Количество');
```

Вывод:
Максимальное количество недвижимости продается с жилой площадью от ~1600 до 2200 кв.м., основная часть недвижимости продается с жилой площадью от ~500 до 6000 кв.м.. Чем больше площадь, тем меньше продаж

```python
plt.figure(figsize=(6, 6))
plt.hist(df['yr_built'], bins = 300)
plt.title('распределение в зависимости от года')
plt.xlabel('Год')
plt.ylabel('Количество');
```

Вывод:
Сиьные Просадки в постройке были примерно в 1930-е и в 1970-е, наибольшее количество домов начало строиться после 2000-х

### Задача 2.1

## Задача 2.2

```python


data2 = df['floors'].value_counts(normalize=True)*100
data2
```

```python
plt.figure(figsize=(6, 6))
plt.pie(data2.values, autopct='%1.0f%%', labels=data2.index)
plt.legend(data2.index);
plt.title('Распределение этажей домов');
```

Вывод: 
Почти 90% домов имеют только 1 или 2 этажа
